```python
df_ = df[~(df.YEAR == 2019)]
df_.head()
```

```python
for i,el in enumerate(df_):
    print(el +'   is')
    print(pd.unique(df_[el]))
    print('-'*30)
```

```python
# generate winrate
df_['WINRATE'] = df_['W']/df_['G']
# generate winrate above bubble
df_['WINRATE_BUBBLE'] = df_['WAB']/df_['G']
#drop W ,G ,team name, Winsabovebubble
df_.drop(columns = ['W','G','TEAM','WAB'],inplace = True)
df_.columns.values
```

```python
#mapping POSTSEASON
pos_mapping = {'Champions':0,'2ND':1,'F4':2,'E8':3,'S16':4,'R32':5,'R64':6,'R68':7}
df_['POSTSEASON'] = df_['POSTSEASON'].map(pos_mapping)
df_['POSTSEASON'] = df_['POSTSEASON'].fillna(8)
```

```python
#mapping SEED, punish team without seed
df_['SEED'] = df_['SEED'].fillna(30)
```

```python
#mapping CONF
conf_mapping = {'WCC':0,'ACC':1 ,'B10':2 ,'SEC':3, 'B12' :4,'Amer':5, 'BE' :6,'MAC':7, 'SC':8, 'MWC':9, 'A10':10, 'P12':11,
 'OVC': 12,'WAC':13, 'BSth':14, 'BW':15, 'AE': 16,'CAA':17, 'Ivy':18, 'Horz':19,'SB':20, 'CUSA':21, 'Pat':22, 'BSky':23,
 'MVC': 24,'Slnd':25, 'Sum' :26,'MAAC': 27,'SWAC':28, 'NEC':29 ,'MEAC':30, 'ASun':31}
df_['CONF'] = df_['CONF'].map(conf_mapping)
df_['CONF'] = df_['CONF'].fillna(100)
```

```python
train_set = df_[~(df_.YEAR == 2018)]
test_set = df_[(df_.YEAR == 2018)]
```
